# Add gender

In the next phase of this script we add gender to the business recreationist dataset based on the firstname of each contender. If you have a lot of patience you can run the entire code or get our <b>genderbynames.csv</b> code from the google drive https://drive.google.com/drive/folders/1PxVVV8jvPLDTuU3Fz6Umo4QN2aJcJJiw

In case you dont want to wait four hour to get all the name run the next lines 

```python
df_bus['firstname'] = df_bus['name'].apply(lambda x : re.match('([^-]+)',re.match('([^\s]+)',str(x))[0])[0].lower().title())
```

```python
df_bus = pd.merge(df_bus,pd.read_csv('..\data\genderbynames.csv').drop(columns = 'Unnamed: 0'),
        how='left',left_on='firstname',right_on='name')

df_bus.to_csv('../data/businessrunnersdtd_16_18.csv')
```
