### EXPORTING DATA TO SQL

 #### Here i want to put all the data that has been collected in csv files to sql database

```python
#importing libraries
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
```

```python
#reading data from csv files
data_recreation = pd.read_csv("damtotdamloopdata.csv").drop(columns=['Unnamed: 0'])
data_business = pd.read_csv("businessdamtotdamloopdata.csv").drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
```

```python
#creating engine to connect to database
engine = create_engine('mysql+pymysql://root:1234@localhost/damtotdamloop')
```

```python
#exporting to sql
data_recreation.to_sql('Recreational_runners', con = engine, if_exists = 'append', chunksize = 1000)
data_business.to_sql('Business_runners', con = engine, if_exists = 'append', chunksize = 1000)
```

```python
#reading data from sql
data_recreation= pd.read_sql_query('SELECT * FROM Recreational_runners', engine)
data_business=   pd.read_sql_query('SELECT * FROM  Business_runners', engine)
```
