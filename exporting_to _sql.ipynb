### EXPORTING DATA TO SQL

 #### Here i want to put all the data that has been collected in csv files to sql database.Data can be downloaded from [here](https://drive.google.com/drive/folders/1DpwzQGl6QMS_rbKJlHM2aj8xFpkcqm9h) and can be placed in the same folder as the notebook to be accessed.

```python
#importing libraries
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine
```

```python
#reading data from csv files
df_bus = pd.read_csv("businessrunnersdtd_16_18 .csv").drop(columns=['Unnamed: 0']).drop_duplicates()
df_recr = pd.read_csv('recreationalrunnersdtd_16_18.csv').drop(columns=['Unnamed: 0'])
```

```python
#creating engine to connect to database
engine = create_engine('mysql+pymysql://root:1234@localhost/damtotdamloop')
```

```python
#exporting to sql
df_recr.to_sql('Recreational_runners', con = engine, if_exists = 'replace', chunksize = 1000)
df_bus.to_sql('Business_runners', con = engine, if_exists = 'replace', chunksize = 1000)
```

```python
#reading data from sql
df_recr= pd.read_sql_query('SELECT * FROM Recreational_runners', engine)
df_bus=   pd.read_sql_query('SELECT * FROM  Business_runners', engine)
```
